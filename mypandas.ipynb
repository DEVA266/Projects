```python

data[(data["Height"] > 200) & (data["Year"] > 2000) & (data["Games"] == "2012 Summer") & (data["City"] == "London") & (data["Sport"] == "Swimming") ]
```

```python
display(data[data["Name"].str.contains("Deva")].head(10)) # checks wheather the given column and that object in here it is str and it contains the character
# we can also use the case system to mention that the given case is doesn't check for the uppercase or lowercase
data[data["Name"].str.contains("Deva" ,case=False)].head(10)
```

```python
data[(data["Name"].str.contains("Deva" , case=False) )& (data["Team"] == "India")]
```
